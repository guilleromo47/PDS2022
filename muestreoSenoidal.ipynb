```python
# -*- coding: utf-8 -*-
"""
Created on Wed Aug 24 21:33:53 2022

@author: Guillermo Daniel Spagnolo
         daniespagnolo@frba.utn.edu.ar
"""

#Sintetizar una senoidal de  1000 muestras
#frecuencia de muestreo 1000Hz
#frecuencia de la senoidal 1Hz
#amplitud unitaria
#fase nula
#valor medio nulo

import numpy as np
import matplotlib.pyplot as plt


N  = 1000 # muestras
fs = 1000 #Hz#frec muestreo


def mi_funcion_sen( vmax , dc , ff, ph, nn , fs ):

    ts = 1/fs
    tt = np.linspace(0,(N-1)*ts,nn) #0 a (N-1)*Ts)

    xx = dc + vmax*np.sin(2*np.pi*ff*tt + ph)
    return (tt, xx)

[tt, xx ]= mi_funcion_sen( vmax = 1, dc = 0, ff = 1, ph=0, nn = N, fs = fs)#senoidal con parametros estandar

plt.figure(1)
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')  
plt.plot(tt,xx,"r")
plt.show()


[tt, yy ]= mi_funcion_sen( vmax = 50, dc = 20, ff = 1, ph=0, nn = N, fs = fs)

plt.figure(2)
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')  
plt.plot(tt,yy,"g")
plt.show()

[tt, zz ]= mi_funcion_sen( vmax = 1, dc = 1000, ff = 501, ph=0, nn = N, fs = fs)#Artefacto y con centro en amplitud 1000

plt.figure(3)
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')  
plt.plot(tt,zz,"b-")
plt.show()


[tt, jj ]= mi_funcion_sen( vmax = 1, dc = 0, ff = 499, ph=240, nn = N, fs = fs)
plt.figure(4)
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')  
plt.plot(tt,jj,"r")
plt.show()

[tt, kk ]= mi_funcion_sen( vmax = 1, dc = 0, ff = 750, ph=0, nn = N, fs = fs)
plt.figure(5)
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')  
plt.plot(tt,kk,"")
plt.show()







```
